### Regresión Lineal

Este modelo sólo está conformado por dos variables estadísticas llamadas $X$ y $Y$. Para la regresión lineal simple, se asume que $X$ y $Y$ se relacionan mediante la relación funcional:

$$Y= \beta_0 + \beta_1 X + \epsilon$$

donde $\beta _{0},\beta _{1}\in \mathbb {R}$ son constantes desconocidas llamadas coeficientes de regresión o parámetros. 

### Estimación de los parámetros

Dado que los parámetros $\beta _{0}$ y $\beta _{1}$ son constantes desconocidas, estas deben estimarse mediante los datos de la muestra, supóngase que se tiene $n$ datos  $\{x_{i},y_{i}\}_{i=1}^{n}$, se estimarán los parámetros utilizando el método de mínimos cuadrados.

Se estiman $\beta _{0}$ y $\beta _{1}$ tal que la suma de los cuadrados de las diferencias entre las observaciones $y_{i}$ y la recta de regresión sea mínima, esto es, buscamos minimizar la función error cuadrático dada por:

$$ S(\beta _{0},\beta _{1})=\sum _{i=1}^{n}\left(y_{i}-\beta _{0}-\beta _{1}x_{i}\right)^{2}$$

La función de error cuadrático $ S(\beta _{0},\beta _{1})$ alcanza un mínimo en el punto $\left({\hat {\beta }}_{0},{\hat {\beta }}_{1}\right)$ tal que:

$$ \left.{\frac {\partial S}{\partial \beta _{0}}}\right|_{{\hat {\beta }}_{0},{\hat {\beta }}_{1}}=\left.{\frac {\partial S}{\partial \beta _{1}}}\right|_{{\hat {\beta }}_{0},{\hat {\beta }}_{1}}=0$$ 

entonces derivando respecto a $ \beta _{0}$ y $ \beta _{1}$, evaluando en ${\hat {\beta }}_{0}$ y $ {\hat {\beta }}_{1}$ e igualando a cero, obtenemos el siguiente sistema de ecuaciones: 


 $$ {\begin{cases}\displaystyle -2\sum _{i=1}^{n}\left(y_{i}-{\hat {\beta }}_{0}-{\hat {\beta }}_{1}x_{i}\right)=0\\\displaystyle -2\sum _{i=1}^{n}\left(y_{i}-{\hat {\beta }}_{0}-{\hat {\beta }}_{1}x_{i}\right)x_{i}=0\end{cases}}$$ 

estas dos ecuaciones son conocidas como ecuaciones normales la solución de dicho sistema de ecuaciones está dada por:

$$ {\begin{aligned}{\hat {\beta }}_{1}&={\frac {\displaystyle \sum _{i=1}^{n}x_{i}y_{i}-n{\bar {x}}{\bar {y}}}{\displaystyle \sum _{i=1}^{n}x_{i}^{2}-{\frac {1}{n}}\left(\sum _{i=1}^{n}x_{i}\right)^{2}}}\\{\hat {\beta }}_{0}&={\frac {\displaystyle \sum _{i=1}^{n}x_{i}^{2}\sum _{i=1}^{n}y_{i}-\sum _{i=1}^{n}x_{i}y_{i}\sum _{i=1}^{n}x_{i}}{\displaystyle n\sum _{i=1}^{n}x_{i}^{2}-\left(\sum _{i=1}^{n}x_{i}\right)^{2}}}={\bar {y}}-{\hat {\beta }}_{1}{\bar {x}}\end{aligned}}$$

La interpretación del parámetro medio ${\beta _{1}}$ es que un incremento en $X_i$ de una unidad, $Y_i$ incrementará en ${\beta _{1}}$. Luego el modelo ajustado de regresión lineal simple es: 

$$ {\hat {y}}={\hat {\beta }}_{0}+{\hat {\beta }}_{1}x $$

## Ejemplo

A continuación, se muestra un ejemplo de regresión lineal.

Primero se generaran unos datos que siguen una línea recta, y se le añade un ruido gaussiano o error. Para esto, se usa la librería de python NumPy. En este caso se generan inicialmente 50 datos de la función: 

$$y= 2.5 +0.2x+N(0,0.3)$$

El código en python quedaría así:

```python
from sklearn.linear_model import LinearRegression #Regresión Lineal con scikit-learn
def f(x):  # función f(x) = 0.1*x + 1.25 + 0.2*Ruido_Gaussiano
    np.random.seed(50) # para poder reproducirlo
    y = 0.2*x + 2.5 + 0.3*np.random.randn(x.shape[0])
    return y
x = np.arange(0, 20, 0.5) # generamos valores x de 0 a 20 en intervalos de 0.5
y = f(x) # calculamos y a partir de la función que hemos generado
# hacemos un gráfico de los datos que hemos generado
plt.scatter(x,y,label='data', color='blue')
plt.title('Datos');
```

A continuaciòn, se estiman los parámetros $\beta_0$ y $\beta_1$.

```python
# Importamos la clase de Regresión Lineal de scikit-learn
from sklearn.linear_model import LinearRegression 
regresion_lineal = LinearRegression() # creamos una instancia de LinearRegression
# instruimos a la regresión lineal que aprenda de los datos (x,y)
regresion_lineal.fit(x.reshape(-1,1), y) 
# vemos los parámetros que ha estimado la regresión lineal
print('𝛽1 = ' + str(regresion_lineal.coef_) + ', 𝛽0 = ' + str(regresion_lineal.intercept_))
# resultado: w = [0.09183522], b = 1.2858792525736682
```
